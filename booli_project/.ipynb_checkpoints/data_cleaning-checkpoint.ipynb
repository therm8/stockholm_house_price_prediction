```python
#removing duplicates
df = df.drop_duplicates()
```

```python
#creating a column for the number of rooms
df['rooms'] = df['living_area'].apply(lambda x: x.split(',')[0] if ',' in x else 0)
df.rooms = df.rooms.astype(str)
```

```python
df['rooms'] = df['rooms'].apply(lambda x: x.strip(' rum'))
```

```python
text = 'Villa, Enskededalen'
l = text.split(' ')
print(l[1])
```

```python
#create column for main area 
df['main_area'] = df['living_area'].apply(lambda x: x.split(' ')[3] if ',' in x else x.split(' ')[1])
```

```python
#create column for alternative area
df['alt_area'] = df['living_area'].apply(lambda x: x.split('+')[1] if '+' in x else 0)
df.alt_area = df.alt_area.astype(str)
df['alt_area'] = df['alt_area'].apply(lambda x: x.replace(' m²', ''))
```

```python
#extracting the location from the district column into new column location
df['location'] = df['district'].apply(lambda x: x.split(' ')[2])
```
